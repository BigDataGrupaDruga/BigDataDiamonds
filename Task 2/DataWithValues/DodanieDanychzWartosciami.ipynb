```python
import pandas as pd

# Link do pliku
url = 'https://raw.githubusercontent.com/BigDataGrupaDruga/BigDataDiamonds/refs/heads/main/CleanedData/WyczyszczoneDane.csv'
dataset = pd.read_csv(url)

df = dataset
```

```python
df.head()
```

```python
cut_mapping = {
    'FAIR': 1,
    'GOOD': 2,
    'VERY GOOD': 3,
    'PREMIUM': 4,
    'IDEAL': 5
}
```

```python
df['cut value'] = df['cut'].map(cut_mapping)
```

```python
df.head()
```

```python
color_mapping = {
    'J': 1,
    'I': 2,
    'H': 3,
    'G': 4,
    'F': 5,
    'E': 6,
    'D': 7
}
```

```python
df['color value'] = df['color'].map(color_mapping)
```

```python
df.head()
```

```python
df.tail()
```

```python
clarity_mapping = {
    'I1': 1,
    'SI2': 2,
    'SI1': 3,
    'VS2': 4,
    'VS1': 5,
    'VVS2': 6,
    'VVS1': 7,
    'IF': 8
}
```

```python
df['clarity value'] = df['clarity'].map(clarity_mapping)
```

```python
df.head(15)
```

```python
df.drop('Unnamed: 0', axis=1, inplace=True)
```
