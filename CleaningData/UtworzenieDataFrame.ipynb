```python
import pandas as pd
import os
import tempfile

# Link do pliku
url = 'https://raw.githubusercontent.com/BigDataGrupaDruga/BigDataDiamonds/refs/heads/main/CleaningData/diamonds_noisy.csv'

dataset = pd.read_csv(url)

# Zapisanie danych do tymczasowego pliku
with tempfile.NamedTemporaryFile(delete=False, mode='w', suffix='.csv') as tmp_file:
    dataset.to_csv(tmp_file.name, index=False)
    tmp_file_path = tmp_file.name

# Wczytanie danych jako zwykły tekst
with open(tmp_file_path, 'r') as file:
    lines = file.readlines()

# Przetworzenie danych
data = []
for line in lines:
    # Usunięcie niepotrzebnych znaków, takich jak `'``
    line = line.replace("'`", '')
    
    # Naprawienie pisowni
    line = line.replace("Go0d", "Good").replace("Pre_AEiu_AE", "Premium")
    
    # Podział na kolumny
    columns = line.strip().split(',')
    data.append(columns)

# Tworzenie DataFrame
df = pd.DataFrame(data)

# Usunięcie tymczasowego pliku
os.unlink(tmp_file_path)
```
