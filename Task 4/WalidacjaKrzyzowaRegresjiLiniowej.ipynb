```python
import pandas as pd

url = 'https://raw.githubusercontent.com/BigDataGrupaDruga/BigDataDiamonds/refs/heads/main/Task%202/DataWithValues/DanezWartosciami.csv'
df = pd.read_csv(url)
```

```python
df.head()
```

```python
df.tail()
```

```python
df.drop(columns=['Unnamed: 0', 'cut', 'color', 'clarity'], inplace=True)
```

```python
df.head()
```

```python
X = df[['carat']]
X.head()
```

```python
Y = df['price']
Y.head()
```

```python
# import modułu Scikit-learn 
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import KFold, cross_val_score

# walidacja krzyżowa
kfold = KFold(n_splits=3, random_state=7, shuffle=True)
model = LinearRegression()
wyniki = cross_val_score(model, X, Y, cv=kfold)
print(wyniki)
print("Uśredniona wartość wyników dla współczynnika R2: %.3f oraz średnia wartość wyników odchylenia standardowego regresji (błędu standardowego estymacji): %.3f" % (wyniki.mean(), wyniki.std()))
```

```python
# Powyżej wykonana została walidacja krzyżowa dla regresji liniowej, analizowanej w osobnym pliku liczby karatów do ceny diamentów.
# Przeprowadzona walidacja krzyżowa pozwalająca na analizę dużego zbioru badanych danych określająca współcznik determinacji (zwany r**2) daje obiecujące wyniki.
# Uśredniona wartość wyników przeprowadzonej walidacji krzyżowej wynosi 0.856, co oznacza dobrą korelację regresji (1.0 - idealnie, 0.0 - fatalnie).
# Oznacza to też, że model z "wytrenowaną" regresją liniową dobrze działa w odniesieniu do zbioru testowego.
```
